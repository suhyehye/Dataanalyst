```python
import pandas as pd
import numpy as np

data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

a = ['sex','smoker','region']
for i in a:
    le = LabelEncoder()
    data[i] = le.fit_transform(data[i])
    
print(data.head())
```

```python
X = data.iloc[:,0:6]
y = data[['charges']]

print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train['age'] = scaler.fit_transform(X_train[['age']])
X_train.head()
```

```python
ss = ['age','bmi']
for s in ss:
    scaler = MinMaxScaler()
    X_train[s] = scaler.fit_transform(X_train[[s]])
    X_test[s] = scaler.transform(X_test[[s]])
    
print(X_train.head())
print(X_test.head())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
pred_train = lr.predict(X_train)
lr.score(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
lr_mse = mean_squared_error(y_train,pred_train)
print("Linear RMSE : ",np.sqrt(lr_mse))
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
pred_train = svr.predict(X_train)
print(svr.score(X_train,y_train))
svr_mse = mean_squared_error(y_train,pred_train)
print("SVR RMSE : ",np.sqrt(svr_mse))
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
tree.fit(X_train,y_train)
pred_train = tree.predict(X_train)
print(tree.score(X_train,y_train))
tree_mse = mean_squared_error(y_train,pred_train)
print("Tree RMSE : ",np.sqrt(tree_mse))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
pred_train = rf.predict(X_train)
print(rf.score(X_train,y_train))
```

```python
model = RandomForestRegressor()
model.fit(X_train,y_train)
pred_test = model.predict(X_test)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':[100,200,300,400],'max_depth':[1,3,5,7,8],'min_samples_leaf':[1,3,5,7]}
grid_search = GridSearchCV(RandomForestRegressor(),param_grid,cv=5)
grid_search.fit(X_train,y_train)

print("Best Parameter : {}".format(grid_search.best_params_))
print("Best Score :{:.4f}".format(grid_search.best_score_))
print("Test Set Score : {:.4f}".format(grid_search.score(X_test,y_test)))
```
