```python
import pandas as pd
import numpy as np
import datetime as dt

b2 = pd.read_csv('basic2.csv',parse_dates = ['Date'])
b2.head()
```

```python
#주말데이터 생성
b2['year'] = b2['Date'].dt.year
b2['month'] = b2['Date'].dt.month
b2['day'] = b2['Date'].dt.day
b2['dayofweek'] = b2['Date'].dt.dayofweek

b2.head()
```

```python
b2['weekend'] = b2['dayofweek'].apply(lambda x:x>=5,1,0)
b2.head()
```

```python
#5월 주말 평균
weekend = b2[(b2['weekend']==True)&(b2['month']==5)&(b2['year']==2022)]['Sales'].mean()
#5월 평일 평균
weekday = b2[(b2['weekend']==False)&(b2['month']==5)&(b2['year']==2022)]['Sales'].mean()

#차이
print(round(abs(weekend-weekday),1))
```
