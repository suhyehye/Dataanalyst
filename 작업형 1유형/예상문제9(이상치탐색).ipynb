```python
import pandas as pd
import numpy as np

b1 = pd.read_csv("basic1.csv")
b1.head()
```

```python
#이상치 탐색(나이가 소수점인것)
b1_ = b1[(b1['age']-np.floor(b1['age'])!=0)]
b1_.head()
```

```python
#이상치를 포함한 데이터의 평균
#올림 
up = np.ceil(b1['age']).mean()

#내림
down = np.floor(b1['age']).mean()

#버림
trunc = np.trunc(b1['age']).mean()
```

```python
#평균값 더해서 출력
print(up+down+trunc)
```
