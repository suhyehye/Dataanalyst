```python
import pandas as pd
import numpy as np

data = pd.read_csv('mtcars.csv')
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = data[['qsec']]
scaler.fit(x)
data['qsec'] = scaler.transform(x)
data.head()
```
