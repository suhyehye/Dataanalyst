```python
import pandas as pd
import numpy as np

data = pd.read_csv('covid_vaccination_vs_death_ratio.csv')
data.head()
```

```python
data2 = data.groupby('country').max()
data2.head()
```

```python
#접종률 상위 10개
top = data2['ratio'].sort_values(ascending=True).iloc[:10].mean()
bottom = data2['ratio'].sort_values(ascending=False).iloc[:10].mean()
```

```python
print(top-bottom)
```
