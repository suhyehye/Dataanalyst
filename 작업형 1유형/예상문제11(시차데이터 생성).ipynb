```python
import pandas as pd
import numpy as np

b2 = pd.read_csv('basic2.csv')
b2.head()
```

```python
#1일 차이가 나는 시차 특성 만들기
b2['previous_PV'] = b2['PV'].shift(1)
b2.head()
```

```python
b2['previous_PV'] = b2['previous_PV'].fillna(method = 'bfill')
```

```python
b2.head()
```

```python
#event가 1이면서 sales가 1000000이하인 컬럼의 합
a = b2[(b2['Events']==1)&(b2['Sales'] <= 1000000)]
a = pd.DataFrame(a)
print(a['previous_PV'].sum())
```
