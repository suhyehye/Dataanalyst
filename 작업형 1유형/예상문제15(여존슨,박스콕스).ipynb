```python
import pandas as pd
import numpy as np

b1 = pd.read_csv('basic1.csv')
b1.head()
```

```python
#20세 이상인 데이터 추출
b1 = b1[b1['age']>20]
b1.shape
```

```python
#f1컬럼 결측치를 최빈값으로 채움
b1['f1'] = b1['f1'].fillna(b1['f1'].mode()[0])
b1.isnull().sum()
```

```python
#yeo-Johnson
from sklearn.preprocessing import power_transform
b1['y'] = power_transform(b1[['f1']])
b1.head()
```

```python
b1['y'] = power_transform(b1[['f1']],standardize=False)
b1.head()
```

```python
#box-cox값
b1['b'] = power_transform(b1[['f1']],method='box-cox')
b1.head()
```

```python
b1['b'] = power_transform(b1[['f1']],method='box-cox',standardize=False)
b1.head()
```

```python
round(sum(np.abs(b1['y']-b1['b'])),2)
```
