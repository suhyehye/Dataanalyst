```python
import pandas as pd
data = pd.read_csv('vote.csv',encoding='utf-8')
data.head()
```

```python
X1 = data[['gender','region']]
XY = data[data.columns[2:]]
```

```python
#범주형 변수 변환
X1['gender'] = X1['gender'].replace([1,2],['male','female'])
X1['region'] = X1['region'].replace([1,2,3,4,5],['Sudo','Chungcheung','Honam','Youngman','Others'])
```

```python
X1.head()
```

```python
#범주형변수 one-hot-encoding 변환

X1_dum = pd.get_dummies(X1)
X1_dum.head()
```

```python
#자료 통합 및 저장
Fvote = pd.concat([X1_dum,XY],axis=1)
Fvote.head()
```

```python
Fvote.to_csv("Fvote.csv")
```
